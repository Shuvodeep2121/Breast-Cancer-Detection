```python
df['diagnosis'] = df['diagnosis'] .map({'M':1,'B':0})
```

```python
from sklearn.model_selection import train_test_split

array = df.values


#Define the independent variable as features.
features = array[:,1:]

#Define the target (dependent) variables as labels.
labels=array[:,0]

#create a train/test split using 30% test size.
features_train, features_test, labels_train, labels_test=train_test_split(features,labels,test_size=0.3,random_state=42)

#check the split printing the shape of each set.
print(features_train.shape, labels_train.shape)
print(features_test.shape, labels_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from time import time

def print_ml_results():
    t0=time()
    
    #Create classifier.
    clf = KNeighborsClassifier()
    
    #Fit the classification on the traning features and labels.
    t0=time()
    clf.fit(features_train, labels_train)
    print("Training time:", round(time()-t0, 3),"s")
    
    #Make predictions.
    t1 =time()
    predictions = clf.predict(features_test)
    print("Predictions time:",round(time()-t1,3),"s")
    
    #Evaluate the model.
    accuracy = clf.score(features_test,labels_test)
    report = classification_report(labels_test,predictions)
    
    #Print the reports
    print("\nReport:\n")
    print("Accuracy: {}".format(accuracy))
    print("\n",report)
    print(confusion_matrix(labels_test,predictions))

print_ml_results()
```
